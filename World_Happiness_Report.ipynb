```python
def region(a,b):
    """ This function creates a list of 1 and 0. 
        The Loop adds 1 if the country is in the region given in the parameter 'b'. 
        a = column with the regions in the dataframe """
    
    c = []
    for reg in a:        
        #return a
        if reg == b:
            c.append(1)
        else:
            c.append(0)
    return c 
W_Europe = region(df['Region'],'Western Europe')
E_Asia = region(df['Region'],'Eastern Asia')
L_America = region(df['Region'],'Latin America and Caribbean')
N_America = region(df['Region'],'North America')
Middle_East_Africa = region(df['Region'],'Middle East and Northern Africa')
Southeastern_Asia = region(df['Region'],'Southeastern Asia')
Australia_Zealand = region(df['Region'],'Australia and New Zealand')
C_E_Europe = region(df['Region'],'Central and Eastern Europe')
Southern_Asia = region(df['Region'],'Southern Asia')
S_S_Africa = region(df['Region'],'Sub-Saharan Africa')

data_h = df.copy()
data_h = data_h.set_index('Country')      # Setting countries as index to create a new data frame with the same index

countries_region = pd.DataFrame(index=data_h.index)      # Creating Data Frame with index countries 



countries_region['Western Europe'] = W_Europe
countries_region['Central and Eastern Europe'] = C_E_Europe
countries_region['Latin America and Caribbean'] = L_America
countries_region['North America'] = N_America
countries_region['Middle East and Northern Africa'] = Middle_East_Africa
countries_region['Southeastern Asia'] = Southeastern_Asia
countries_region['Eastern Asia'] = E_Asia
countries_region['Australia and New Zealand'] = Australia_Zealand
countries_region['Central and Eastern Europe'] = C_E_Europe
countries_region['Southern Asia'] = Southern_Asia
countries_region['Sub-Saharan Africa'] = S_S_Africa
```
